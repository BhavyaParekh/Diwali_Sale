## EXPLORATORY DATA ANALYSIS

### GENDER

```python
df.columns
```

```python
a = sns.countplot(x='Gender',data=df)
```

```python
a = sns.countplot(x='Gender',data=df)
for i in a.containers:
    a.bar_label(i)
```

```python
df.groupby(['Gender'] , as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
```

```python
b=df.groupby(['Gender'] , as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Gender',y='Amount',data=b)
```

#### female purchase more than men

# Age

```python
df.columns
```

```python
 sns.countplot(x='Age Group',data=df , hue='Gender')
```

```python
a = sns.countplot(x='Age Group',data=df , hue='Gender')
for i in a.containers:
    a.bar_label(i)
```

```python
# total amount vs age group
b=df.groupby(['Age Group'] , as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=b)
```

### from above graph we can come to know that age most of buyers are female and there ages are between 26-35

# States

```python
# total number of orders from top cities
```

```python
df.columns
```

```python
a = df.groupby(['State'] , as_index=False)['Orders'].sum().sort_values(by='Orders' , ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=a , x='State',y='Orders')
```

```python
# totalamounts from top cities
```

```python
a = df.groupby(['State'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=a , x='State',y='Amount')
```

### as we can here top cities who have ordered most and total amount spend by top cities

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
```

```python
a = sns.countplot(data=df , x='Occupation')
for i in a.containers:
    a.bar_label(i)
```

```python
b = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=b,x='Occupation',y='Amount')
```

### from above graph we can conclude that most of the buyers are working in IT and Health Care sector

# Product Category

```python
sns.set(rc={'figure.figsize':(24,5)})
a=sns.countplot(data=df,x='Product_Category')
for i in a.containers:
    a.bar_label(i)
```

```python
b = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
sns.barplot(data=b,x='Product_Category',y='Amount')
```

### from above graph we can see that most of order is done in clothing but the amount spend is on food item 

```python
# top most sold products
b = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head()
sns.barplot(data=b,x='Product_ID',y='Orders')
```
